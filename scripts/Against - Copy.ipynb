```python
# read documents
#f=open("All data 01-01-2019 to 02-15-2020 UTF-8.txt",'r', encoding="utf8")
#f=open("reviewes.txt",'r', encoding="utf8")
# get a set of stopwords
stopwords = set(nltk.corpus.stopwords.words("english"))

# create a regex tokenizer that only consider alphabets as token
# this also removes numbers
tokenizer = nltk.tokenize.RegexpTokenizer(r"[a-z]+")

# initialize a stemmer
stemmer = nltk.stem.PorterStemmer()

#Define tokenizer
def tokenize(doc):
    tokens = []
    for token in tokenizer.tokenize(doc):
        if token not in stopwords:
            #tokens.append(stemmer.stem(token))
            tokens.append(token)
            #tokens.append(token)
    return tokens

#define stemmer
def stem(text):
    tokens = []
    for token in tokenizer.tokenize(text):
        if token not in stopwords:
            #tokens.append(stemmer.stem(token))
            tokens.append(token)
            #tokens.append(token)
    return tokens
#Read reviews

AppReviewsList=[]
count=0
for line in df['Full Text']:
  appReviews=line.replace('\n','')
  appReviews=appReviews.lower()
  stemmed= ' '.join(stem(appReviews)) 
  AppReviewsList.append(stemmed) 
  count+=1  
#f.close()  

print (stem("readings"))
print (count)
#print AppReviewsList[29]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
# vectorize
#vectorizer = TfidfVectorizer(stop_words="english")
vectorizer = CountVectorizer(ngram_range=(2,3), stop_words="english")
corpus_vect = vectorizer.fit_transform(AppReviewsList)
print ("Vocaublary size: %d" % len(vectorizer.vocabulary_))

df_vect = pd.DataFrame(corpus_vect.toarray(), columns=vectorizer.get_feature_names())
df_vect
```

```python
# convert the vectorized data to a gensim corpus object
corpus = gensim.matutils.Sparse2Corpus(corpus_vect, documents_columns=False)
# maintain a dictionary for index-word mapping
id2word = dict((v, k) for k, v in vectorizer.vocabulary_.items())
```

```python
# build the lda model
lda = LdaModel(corpus, num_topics=50,id2word=id2word, passes=10)
#print lda.print_topics()

#Print all the topics
for i in range(0, lda.num_topics):
    print (lda.print_topic(i))
    print("-------------------------------------------------------------")
```

```python
# use the lda model to transform documents
lda_docs = lda[corpus]
for row in lda_docs:
    print row
```

```python
# extract the scores and round them to 3 decimal places
#scores = np.round([[doc[1] for doc in row] for row in lda_docs], 3)
scores = [[doc[1] for doc in row] for row in lda_docs]
scores
```

```python
# convert the documents scores into a data frame
df_lda = pd.DataFrame(scores, columns=["topic 1", "topic 2", "topic 3", "topic 4", "topic 5", "topic 6", "topic 7", "topic 8", "topic 9", "topic 10", "topic 11", "topic 12", "topic 13", "topic 14", "topic 15", "topic 16", "topic 17", "topic 18", "topic 19", "topic 20", "topic 21", "topic 22", "topic 23", "topic 24", "topic 25", "topic 26", "topic 27", "topic 28", "topic 29", "topic 30"])
df_lda
```
